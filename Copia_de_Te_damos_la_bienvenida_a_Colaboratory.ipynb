```
import networkx as nx
import matplotlib.pyplot as plt

g_corrected_no_common_to_initial = nx.DiGraph()


# Definir nodos según la lógica corregida
nodes_final_corrected_v2 = {
    "Problema": {"type": "problem"},                  # Nodo inicial
    "Solución Fácil": {"type": "solution"},
    "Solución Difícil": {"type": "solution"},
    "Nodo Solución Común": {"type": "common_solution"},   # Nodo común
    "Subsolución Fácil": {"type": "subsolution"},
    "Subsolución Difícil": {"type": "subsolution"},
}


# Agregar nodos al grafo
for node, attrs in nodes_final_corrected_v2.items():
    g_corrected_no_common_to_initial.add_node(node, **attrs)


# Definir conexiones entre nodos
edges_final_corrected_v2 = [
    # Conexiones desde el problema a soluciones iniciales
    ("Problema", "Solución Fácil"),
    ("Problema", "Solución Difícil"),

    # Conexión de subsoluciones cruzadas (no hacia soluciones iniciales)
    ("Subsolución Fácil","Nodo Solución Común"),
    ("Subsolución Difícil","Nodo Solución Común"),
    ("Solución Difícil","Subsolución Fácil"),
    ("Solución Fácil","Subsolución Difícil"),
]


# Agregar conexiones al grafo
g_corrected_no_common_to_initial.add_edges_from(edges_final_corrected_v2)


# Visualizar el grafo
pos_corrected_v2 = nx.spring_layout(g_corrected_no_common_to_initial)  # Layout para organizar los nodos
plt.figure(figsize=(14, 12))


# Dibujar nodos con colores específicos según su tipo
nx.draw_networkx_nodes(
    g_corrected_no_common_to_initial,
    pos_corrected_v2,
    node_size=1200,
    node_color=[
        "orange" if g_corrected_no_common_to_initial.nodes[n]["type"] == "problem" else
        "lightgreen" if g_corrected_no_common_to_initial.nodes[n]["type"] == "solution" else
        "yellow" if g_corrected_no_common_to_initial.nodes[n]["type"] == "common_solution" else
        "skyblue" if g_corrected_no_common_to_initial.nodes[n]["type"] == "subsolution" else "red"
        for n in g_corrected_no_common_to_initial.nodes
    ]
)


# Dibujar etiquetas de nodos
nx.draw_networkx_labels(g_corrected_no_common_to_initial, pos_corrected_v2, font_size=10, font_color="black")


# Dibujar conexiones/aristas
nx.draw_networkx_edges(g_corrected_no_common_to_initial, pos_corrected_v2, arrowstyle="->", arrowsize=20, edge_color="black")


# Configuración del título
plt.title("Modelo del Grafo Corregido - Solución Común No Conectada a Iniciales")
plt.show()
```

## <strong>Introducción</strong>
</div>

El documento que estás leyendo no es una página web estática, sino un entorno interactivo denominado <strong>notebook de Colab</strong>, que permite escribir y ejecutar código.

Por ejemplo, esta es una <strong>celda de código</strong> con una secuencia de comandos Python corta que calcula un valor, lo almacena en una variable y devuelve el resultado:

```
import networkx as nx
import matplotlib.pyplot as plt

# Definir los nodos en dos conjuntos bipartitos
U = ["Problema", "Solución Fácil", "Solución Difícil"]  # Conjunto de generadores
V = ["Subsolución Fácil", "Subsolución Difícil", "Nodo Solución Común"]  # Conjunto de procesadores

# Crear el grafo dirigido
g_bipartite = nx.DiGraph()

# Agregar nodos con su atributo bipartito
for node in U:
    g_bipartite.add_node(node, bipartite=0)  # Conjunto U
for node in V:
    g_bipartite.add_node(node, bipartite=1)  # Conjunto V

# Definir las conexiones (aristas) entre los conjuntos
edges = [
    # Conexiones del nodo problema a soluciones
    ("Problema", "Solución Fácil"),
    ("Problema", "Solución Difícil"),

    # Conexiones de soluciones a subsoluciones contrarias
    ("Solución Fácil", "Subsolución Difícil"),
    ("Solución Difícil", "Subsolución Fácil"),

    # Conexiones de subsoluciones al nodo común
    ("Subsolución Fácil", "Nodo Solución Común"),
    ("Subsolución Difícil", "Nodo Solución Común"),
]

# Agregar las aristas al grafo
g_bipartite.add_edges_from(edges)

# Visualización del grafo
pos = nx.spring_layout(g_bipartite)  # Layout para visualización
plt.figure(figsize=(10, 7))
nx.draw(
    g_bipartite,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=3000,
    font_size=10,
    font_weight="bold",
    edge_color="gray",
    arrowsize=20
)
plt.title("Flujo de Información Basado en un Grafo Bipartito", fontsize=14)
plt.show()
```

## Ciencia de datos
</div>

Con Colab, puedes aprovechar por completo las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código que se incluye a continuación usa <strong>NumPy</strong> para generar algunos datos aleatorios y <strong>matplotlib</strong> para visualizarlos. Para editar el código, haz clic en la celda y comienza a editar.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
